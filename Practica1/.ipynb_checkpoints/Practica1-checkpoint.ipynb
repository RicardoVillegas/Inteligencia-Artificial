```python
#Practica con DFS
class Graph:
    def __init__(self):
        self.nodos = {}

    def laberinto_to_graph(self, r_cuadrícula_size, c_cuadrícula_size, pacman_r, pacman_c, comida_r, comida_c, laberinto):
        empty_spaces = []
        bordes = []
        self.start = (pacman_r,pacman_c)  #Valor de inicio
        self.meta = (comida_r,comida_c)       #Meta
        for r,line in enumerate(laberinto):
            for c,char in enumerate(line):
                vertice = (r,c)
                if char == '%':  # Si en los casos de laberinto ahi %
                    continue
                if r<r_cuadrícula_size-1 and laberinto[r+1][c] != '%':    #realiza estas funciones en caso de que en el
                    self.add_borde((vertice,(r+1,c)))               #laberinto se encuente un %
                if c<c_cuadrícula_size-1 and laberinto[r][c+1] != '%':
                    self.add_borde((vertice,(r,c+1)))

    def add_borde(self, vertice):
        vertice1,vertice2 = vertice
        borde = []
        if vertice1 in self.nodos:
            borde = self.nodos[vertice1]
            borde.append(vertice2)
        else:
            borde.append(vertice2)
            self.nodos[vertice1] = borde
        borde = []
        if vertice2 in self.nodos:
            borde = self.nodos[vertice2]
            borde.append(vertice1)
        else:
            borde.append(vertice1)
            self.nodos[vertice2] = borde

    def get_camino(self,padre, start, meta):   #obtener camino
        camino = [meta]
        while camino[-1] != start:
            camino.append(padre[camino[-1]])
        camino.reverse()
        return camino

    def nodos_ordenar(self):
        for node in self.nodos:
            list = self.nodos[node]
            self.ordenar(list)
   
 #Funcion para ordenar los nodos vecino segun HackerRank
    '''
    Jerarquia:
    1.-Arriba
    2.-Izquierda
    3.-Derecha
    4.-Abajo
    '''

    def ordenar(self, list):
        for i in range(0,len(list)):
            for j in range(i+1,len(list)):
                if list[i][0] < list[j][0]:
                    continue
                elif list[i][0] == list[j][0]:
                    if list[i][1] < list[j][1]:
                        continue
                    else:
                        burbuja = list[i]
                        list[i] = list[j]
                        list[j] = burbuja
                else:
                    burbuja = list[i]
                    list[i] = list[j]
                    list[j] = burbuja

    def DFSsearch(self, vertice,visited): #Algoritmo de busqueda DFS
        frontera = [self.start]
        visited = []
        padre = {}
        while frontera:
            estado = frontera.pop()
            visited.append(estado)
            if estado == self.meta:
                print(len(visited))
                for v in visited:
                    x,y = v
                    print(str(x) + " " + str(y))
                camino = self.get_camino(padre,self.start,self.meta)
                print(len(camino)-1)
                for v in camino:
                    x,y = v
                    print(str(x) + " " + str(y))
                return True
            for vecino in self.nodos[estado]:
                if vecino not in frontera and vecino not in visited:
                    padre[vecino] = estado
                    frontera.append(vecino)

def dfs( r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula): #implementacion de dfs
    graph = Graph()
    graph.laberinto_to_graph(r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula)
    visited_nodos = []
    start = (pacman_r,pacman_c)
    graph.nodos_ordenar()
    graph.DFSsearch(start,visited_nodos)
    return

readfile = True
cuadrícula = []
if readfile:
    text = open("resources/laberinto/laberinto1.txt","r")   #Se lee el documento desde una ruta
    pacman_r, pacman_c = [ int(i) for i in text.readline().strip().split() ]
    comida_r, comida_c = [ int(i) for i in text.readline().strip().split() ]
    r,c = [ int(i) for i in text.readline().strip().split() ]    #Se lleva acabo el proceso de lectura de datos
    for i in range(0, r):
        cuadrícula.append(text.readline().strip())
else:
    pacman_r, pacman_c = [ int(i) for i in input().strip().split() ]
    comida_r, comida_c = [ int(i) for i in input().strip().split() ]
    r,c = [ int(i) for i in input().strip().split() ]
    for i in range(0, r):
        cuadrícula.append(input().strip())

dfs(r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula)
```

```python
#Practica con BFS
from collections import deque  #cola

class Graph:
    def __init__(self):
        self.nodos = {}

    def laberinto_to_graph(self, r_cuadrícula_size, c_cuadrícula_size, pacman_r, pacman_c, comida_r, comida_c, laberinto):
        empty_spaces = []
        bordes = []
        self.start = (pacman_r,pacman_c) #Valores iniciales
        self.meta = (comida_r,comida_c)      #Meta a conseguir
        for r,line in enumerate(laberinto):
            for c,char in enumerate(line):
                vertice = (r,c)
                if char == '%':      # Si en los casos de laberinto ahi %
                    continue
                if r<r_cuadrícula_size-1 and laberinto[r+1][c] != '%':
                    self.add_borde((vertice,(r+1,c)))             #realiza estas funciones en caso de que en el
                if c<c_cuadrícula_size-1 and laberinto[r][c+1] != '%':  #laberinto se encuente un % 
                    self.add_borde((vertice,(r,c+1)))

    def add_borde(self, vertice):  #Anadir borde
        vertice1,vertice2 = vertice
        borde = []
        if vertice1 in self.nodos:
            borde = self.nodos[vertice1]
            borde.append(vertice2)
        else:
            borde.append(vertice2)
            self.nodos[vertice1] = borde
        borde = []
        if vertice2 in self.nodos:
            borde = self.nodos[vertice2]
            borde.append(vertice1)
        else:
            borde.append(vertice1)
            self.nodos[vertice2] = borde

    def get_camino(self,padre, start, meta): #Camino 
        camino = [meta]
        while camino[-1] != start:
            camino.append(padre[camino[-1]])
        camino.reverse()
        return camino

    def BFSsearch(self, start):  #Algoritmo de busqueda BFS
        frontera = deque([start])    
        explored = []
        padre = {}
        while frontera:
            estado = frontera.popleft()
            explored.append(estado)
            if estado == self.meta:
                print(len(explored))
                for node in explored:
                    x,y = node
                    print(str(x) + " " + str(y))
                camino = self.get_camino(padre, self.start, self.meta)
                print(len(camino)-1)
                for node in camino:
                    x,y = node
                    print(str(x) + " " + str(y))
                return True
            for vecino in self.nodos[estado]:
                if vecino not in frontera and vecino not in explored:
                    padre[vecino] = estado
                    frontera.append(vecino)

def nextMove( r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula):
    graph = Graph()
    graph.laberinto_to_graph(r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula)
    start = (pacman_r,pacman_c)
    graph.BFSsearch(start)

readfile = True
cuadrícula = []
if readfile:
    text = open("resources/laberinto/laberinto2.txt","r")  #Se lee el documento desde una ruta
    pacman_r, pacman_c = [ int(i) for i in text.readline().strip().split() ] 
    comida_r, comida_c = [ int(i) for i in text.readline().strip().split() ]  #Se lleva acabo el proceso de lectura de datos
    r,c = [ int(i) for i in text.readline().strip().split() ]
    for i in range(0, r):
        cuadrícula.append(text.readline().strip())
else:
    pacman_r, pacman_c = [ int(i) for i in input().strip().split() ]
    comida_r, comida_c = [ int(i) for i in input().strip().split() ]
    r,c = [ int(i) for i in input().strip().split() ]
    for i in range(0, r):
        cuadrícula.append(input().strip())

nextMove(r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula)

'''
Used for camino finding:
https://stackoverflow.com/questions/8922060/how-to-trace-the-camino-in-a-breadth-first-search
'''
```

```python
#Practica con A*
from collections import deque

class Graph:
    def __init__(self):
        self.nodos = {}

    def laberinto_to_graph(self, r_cuadrícula_size, c_cuadrícula_size, pacman_r, pacman_c, comida_r, comida_c, laberinto):
        self.start = (pacman_r,pacman_c) #Valores de inicio
        self.meta = (comida_r,comida_c)      #Meta de pacman
        for r,line in enumerate(laberinto):
            for c,char in enumerate(line):
                vertice = (r,c)
                if char == '%': # Si en los casos de laberinto ahi %
                    continue
                if r<r_cuadrícula_size-1 and laberinto[r+1][c] != '%': #realiza estas funciones en caso de que en el
                    self.add_borde((vertice,(r+1,c)))            #laberinto se encuente un %
                if c<c_cuadrícula_size-1 and laberinto[r][c+1] != '%':
                    self.add_borde((vertice,(r,c+1)))

    def add_borde(self, vertice): 
        vertice1,vertice2 = vertice
        borde = []
        if vertice1 in self.nodos:
            borde = self.nodos[vertice1]
            borde.append(vertice2)
        else:
            borde.append(vertice2)
            self.nodos[vertice1] = borde
        borde = []
        if vertice2 in self.nodos:
            borde = self.nodos[vertice2]
            borde.append(vertice1)
        else:
            borde.append(vertice1)
            self.nodos[vertice2] = borde

    def get_camino(self,padre, start, meta):
        camino = [meta]
        while camino[-1] != start:
            camino.append(padre[camino[-1]])
        camino.reverse()
        return camino

    def deleteMin(self, list, dest_node):
        #Funcion que regresa el nodo con menor coste segun la siguiente formula: Cost = d(s,c) + h(c)
        min_cost = self.manhattan_distancia(list[0], dest_node)
        min_item = list[0]
        for item in list:
            cost = self.manhattan_distancia(item,dest_node)
            if cost < min_cost:
                min_cost = cost
                min_item = item
        return list.pop(list.index(min_item))

    def manhattan_distancia(self,source,dest_node): #Distancia Manhattan Implementacion
        return abs(source[0] - dest_node[0]) + abs(source[1] - dest_node[1])

    def ASTARsearch(self, start): #Algoritmo de busqueda A*
        frontera = [start]
        explored = []
        padre = {}
        while frontera:
            estado = self.deleteMin(frontera, self.meta)
            explored.append(estado)
            if estado == self.meta:
                camino = self.get_camino(padre,self.start,self.meta)
                print(len(camino)-1)
                for v in camino:
                    x,y = v
                    print(str(x) + " " + str(y))
                return True
            for vecino in self.nodos[estado]:
                if vecino not in frontera and vecino not in explored:
                    padre[vecino] = estado
                    frontera.append(vecino)
                #vecino in frontera:
                    #frontera.decreaseKey(vecino)

def a_star( r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula):
    graph = Graph()
    graph.laberinto_to_graph(r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula)
    start = (pacman_r,pacman_c)
    graph.ASTARsearch(start)

readfile = True
cuadrícula = []
if readfile:
    text = open("resources/laberinto/laberinto4.txt","r") #Se lee el documento desde una ruta
    pacman_r, pacman_c = [ int(i) for i in text.readline().strip().split() ]
    comida_r, comida_c = [ int(i) for i in text.readline().strip().split() ] #Se lleva acabo el proceso de lectura de datos
    r,c = [ int(i) for i in text.readline().strip().split() ]
    for i in range(0, r):
        cuadrícula.append(text.readline().strip())
else:
    pacman_r, pacman_c = [ int(i) for i in input().strip().split() ]
    comida_r, comida_c = [ int(i) for i in input().strip().split() ]
    r,c = [ int(i) for i in input().strip().split() ]
    for i in range(0, r):
        cuadrícula.append(input().strip())

a_star(r, c, pacman_r, pacman_c, comida_r, comida_c, cuadrícula)
```
